```python
import torch
import pandas as pd

from torch.utils.data import Dataset, DataLoader, random_split
from transformers import TrOCRProcessor
from PIL import Image
```

```python
csv_path = "/content/drive/MyDrive/data/updated_labels.csv"

df = pd.read_csv(csv_path)

print("Total Images :", len(df))
print("Missing Images :", (~df["image"].apply(__import__("os").path.exists)).sum())
print("Empty Labels :", df["text"].isna().sum())
```

```python
processor = TrOCRProcessor.from_pretrained(
    "microsoft/trocr-base-printed"
)

print(" Processor Loaded Successfully!")
```

```python
class UrduOCRDataset(Dataset):

    def __init__(self, csv_path, processor):

        self.data = pd.read_csv(csv_path)
        self.processor = processor

        print(f"Dataset Loaded: {len(self.data)} samples")

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):

        row = self.data.iloc[idx]

        image = Image.open(row["image"]).convert("RGB")

        text = row["text"]

        if pd.isna(text):
            text = ""

        text = str(text)

        encoding = self.processor(
            image,
            return_tensors="pt"
        )

        pixel_values = encoding.pixel_values.squeeze(0)

        labels = self.processor.tokenizer(
            text,
            padding="max_length",
            truncation=True,
            max_length=128,
            return_tensors="pt"
        ).input_ids.squeeze(0)

        return {
            "pixel_values": pixel_values,
            "labels": labels
        }
```

```python
dataset = UrduOCRDataset(
    "/content/drive/MyDrive/data/updated_labels.csv",
    processor
)

print("Dataset Size:", len(dataset))
```

```python
sample = dataset[0]

print("Pixel Values Shape:", sample["pixel_values"].shape)
print("Labels Shape:", sample["labels"].shape)
```

```python
train_size = int(0.8 * len(dataset))
test_size = len(dataset) - train_size

train_dataset, test_dataset = random_split(
    dataset,
    [train_size, test_size]
)

print("Training Samples:", len(train_dataset))
print("Testing Samples:", len(test_dataset))
```

```python
train_loader = DataLoader(
    train_dataset,
    batch_size=4,
    shuffle=True
)

test_loader = DataLoader(
    test_dataset,
    batch_size=4,
    shuffle=False
)

print("Train Batches:", len(train_loader))
print("Test Batches:", len(test_loader))
```
