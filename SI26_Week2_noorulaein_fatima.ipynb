```python
# Urdu OCR Project – Week 2

This notebook preprocesses Urdu text images for OCR tasks.
It converts images to grayscale, resizes them, removes noise, and applies binarization.
The notebook also tests Tesseract OCR on Urdu images and evaluates its performance.
```

```python
!pip install opencv-python-headless pillow

import cv2
import numpy as np
from PIL import Image
import os
import matplotlib.pyplot as plt

print('Libraries loaded successfully!')
```

```python
import cv2
import os

def preprocess_image(image_path, save_path):
    # Load image
    img = cv2.imread(image_path)

    if img is None:
        print(f'Could not load: {image_path}')
        return None

    # Step 1: Convert to grayscale (removes colour noise)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Step 2: Resize to standard size
    resized = cv2.resize(gray, (512, 128))

    # Step 3: Remove noise
    denoised = cv2.fastNlMeansDenoising(resized, h=10)

    # Step 4: Binarise (black and white image)
    _, binary = cv2.threshold(
        denoised, 127, 255, cv2.THRESH_BINARY
    )

    # Save processed image
    cv2.imwrite(save_path, binary)

    return binary

# Create output folder in Google Drive
os.makedirs('/content/drive/MyDrive/data/processed', exist_ok=True)

print('Preprocessing function ready!')
```

```python
import glob
import os

# Find all images inside Google Drive data/raw and its subfolders
all_images = glob.glob('/content/drive/MyDrive/data/raw/**/*.jpg', recursive=True)
all_images += glob.glob('/content/drive/MyDrive/data/raw/**/*.png', recursive=True)

print(f'Found {len(all_images)} images to process')

processed_count = 0

for img_path in all_images:
    filename = os.path.basename(img_path)
    save_path = f'/content/drive/MyDrive/data/processed/{filename}'

    result = preprocess_image(img_path, save_path)

    if result is not None:
        processed_count += 1

print(f'Done! Processed {processed_count} images')
print('Check /content/drive/MyDrive/data/processed/ folder')
```

```python
# Install Tesseract Urdu support
!apt-get install -y tesseract-ocr tesseract-ocr-urd
!pip install pytesseract

import pytesseract
from PIL import Image
import glob

# Find both JPG and PNG processed images
test_images = glob.glob('/content/drive/MyDrive/data/processed/*.jpg')
test_images += glob.glob('/content/drive/MyDrive/data/processed/*.png')

# Select first 5 images
test_images = test_images[:5]

print('=== Tesseract Results on Urdu Images ===')
print()

for img_path in test_images:
    img = Image.open(img_path)

    # Use Urdu language model
    result = pytesseract.image_to_string(img, lang='urd')

    print(f'Image: {img_path}')
    print(f'Tesseract output: {result}')
    print('---')
```
